## Import

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Specification

```python
%matplotlib inline
```

```python
x = np.linspace(-10,10, 50)
y = x ** 2

plt.figure(figsize=(15,8))
plt.plot(x, y, c = 'r', ls= '-', lw = 3, marker= 'o', markersize= 15, markeredgecolor='b', markerfacecolor = 'm',
         label = '$y = x^2$')

plt.legend(fontsize= 15, loc= 'upper center' )
plt.grid(c= 'k', ls='--', alpha=0.5)

plt.xlabel('X range', fontsize = 15, fontweight= 2, fontname= 'Times New Roman')
plt.ylabel('Y range',fontsize = 15, fontweight= 2, fontname= 'Times New Roman')
plt.title('Function $Y=X^2$\n in range (-10,10)',fontsize = 20, fontweight= 2, fontname= 'Times New Roman',
          c='darkblue')

#plt.text(-10,110,'first text\nin figure', fontsize = 15, fontweight= 2, fontname= 'Times New Roman')

#plt.ylim(top=120)
#plt.xlim(left=0)

plt.axhline(40, c='k', ls='--')
plt.axvline(0, c='g', ls='--')

#plt.savefig('text_1.jpg', dpi=700)
plt.show()
```

```python
x = np.linspace(-10,10, 50)
y = x ** 2

plt.figure(figsize=(15,8))
plt.plot(x, 10*y,  c = 'r', ls= '-', lw = 3, label = '$y = 10 \cdot x^2$')
plt.plot(x, x**3,c = 'b', ls= '--', lw = 2, label = '$y = x^3$')
plt.plot(x, 100*np.sin(x),c = 'g', ls= '-.', lw = 2, label = '$y = 100 \cdot sin(x)$')

plt.legend(fontsize=15)
plt.grid()
```

```python
plt.figure(figsize=(15,8))
plt.subplot(2,3,1)
plt.plot(x, x**2)

plt.subplot(2,3,2)
plt.plot(x, x**3)

plt.subplot(2,3,3)
plt.plot(x, np.sin(x))

plt.subplot(2,3,5)
plt.plot(x, np.tan(x))

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,8))

ax.plot(x,x**2, label= 'ONE')
ax.plot(x, 50*np.sin(x),label='TWO')
ax.grid()

ax.legend()

ax.set_xlabel('x range')
ax.set_ylabel('y range')
ax.set_title('compare')
```

```python
fig, (ax1, ax2) = plt.subplots(2,1,figsize=(15,8), sharex=True)

ax1.plot(x,x**2, label= 'one')
ax2.plot(x, 50*np.sin(x),label='TWO')
ax1.grid()
ax2.grid()

ax1.legend()
ax2.legend()

ax2.set_xlabel('x range')
ax1.set_ylabel('y range')
ax2.set_ylabel('y range')
ax1.set_title('compare')
```

```python
x = np.linspace(0,100, 100)
y_1 = 20 + x
y_2 = 100 - 2 * x

plt.figure(figsize=(15,8))
plt.plot(x,y_1, c= 'r', lw=2, label= 'Supply')
plt.plot(x,y_2,c= 'b', lw=2, label= 'Demad')

plt.plot(x,y_2 -20,c= 'g', lw=2, label= 'demand after taxation',ls='-.')

plt.xlim(left=0, right=110)
plt.ylim(bottom=0)

plt.axhline(80/3+20, c='k', ls='--',lw=1.5)
plt.axvline(80/3,  c='k', ls='--',lw=1.5)

plt.legend(fontsize=15)
plt.grid()

plt.xlabel('quantity')
plt.ylabel('price')
plt.title('Demand and supply plot')


plt.fill_between(x,y_2, y_1, where= (y_2>y_1) & (x>20), color='darkorange',alpha=0.7)
```

## Bar

```python
plt.figure(figsize=(15,8))
plt.bar([1,3,5,7],[5,8,2,6], label='one')
plt.bar([2,4,6,8],[1,4,8,3], label='two')

plt.legend(fontsize=15, loc= 'upper left')
plt.show()
```

## Scatter

```python
x = np.linspace(0,10,100)
e = np.random.rand(100) *100
y = x + e
scale = np.random.rand(100)*1000
color = np.random.rand(100)

plt.figure(figsize=(15,8))
plt.scatter(x, y, s= scale, c= color)
plt.grid()

plt.colorbar().set_label('color bar', fontsize=15)
plt.xlabel('GDP')
```

```python
z = np.arange(100)
d = np.arange(100)*2
m = np.arange(100)
a = np.random.randn(100)
b = z + a * 20

plt.figure(figsize=(15,8))
plt.scatter(z, b, c=m, s = d)
plt.colorbar()
plt.grid()
plt.show()
```

## Histogram

```python
data = np.random.normal(0,1,100000)

x = np.linspace(-4,4,100)
y = 1/np.sqrt(2*np.pi)* np.exp(-x**2/2)

plt.figure(figsize=(15,8))
plt.plot(x,y, lw=3, c='k')
plt.hist(data, 50, density=1);
```
